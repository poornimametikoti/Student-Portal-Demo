```python
# STEP 1: Upload file
from google.colab import files
uploaded = files.upload()

# STEP 2: Extract filename
filename = list(uploaded.keys())[0]

# STEP 3: Load CSV file using pandas
import pandas as pd
df = pd.read_csv("customer_feedback_sentiment.csv")

print("File loaded successfully!")

# STEP 4: (Optional) Example cleaning
# Remove duplicate rows
df = df.drop_duplicates()

# Remove missing values
df = df.dropna()

# STEP 5: Save cleaned CSV
cleaned_filename = "cleaned_" + filename
df.to_csv(cleaned_filename, index=False)

# STEP 6: Download cleaned CSV
files.download(cleaned_filename)
```

```python
# Install required library
!pip install -q textblob
import nltk
nltk.download('punkt')

# Import library
from textblob import TextBlob

print("Sentiment Analyzer Ready ✅")
print("Type 'exit' to stop.\n")

while True:
    feedback_text = input("Enter customer feedback: ").strip()

    if feedback_text.lower() == "exit":
        print("Program stopped.")
        break

    if feedback_text == "":
        print("⚠️ Please enter some feedback.\n")
        continue

    blob = TextBlob(feedback_text)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        sentiment = "Positive 😊"
    elif polarity < 0:
        sentiment = "Negative 😞"
    else:
        sentiment = "Neutral 😐"

    print("🎯 Sentiment:", sentiment)
    print("📊 Polarity Score:", polarity)
    print("-" * 40)
```
